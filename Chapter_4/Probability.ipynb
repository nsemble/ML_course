### Random Experiment



In particular, a random experiment is a process by which we observe something uncertain. If in an experiment all the possible outcomes are known in advance  and none of the outcomes can be  predicted with certainty, then such an experiment is called a random experiment.Before rolling a die you do not know the result. This is an example of a random experiment.
1.Each performance of random experiment is called as trial.
2.The result of the trial is called as outcome.
3.Set of all possible outcomes is called the sample space.
4.The elements of the sample space are called elementary events.

Example -  Random experiment: toss a coin; sample space: S={heads,tails}


### Events

In probability, the subset of the sample space (or set of outcomes) from an experiment is
known as an Event. For example, if you conduct an experiment by tossing a coin, so when
the coin lands tails, an event can be said to have occurred. The ultimate aim is to find the probability of a certain event.

##### Types of Events

#### Simple Event

If the event E has only one sample point of a sample space, it is called a simple
event or an Elementary Event. It is an event that consists of exactly one outcome.

Example -  Choosing a card from a deck of 52 cards and getting a certain card.

#### Compound Events

As opposed to a simple event, if there is more than one sample point on a sample
space, such an event is called Compound Event. It involves combining two or
more events together and finding the probability of such a combination of events.

Example -  Choosing a card from a deck of 52 cards and getting any heart card.


#### Sure Event

An event which is sure to occur in any given experiment is a certain event. The
probability of this type of event is 1.

Example  - In the below picture, probability that batsman is out is 1 that is it is a sure event.

#### Impossible Event

On the other hand, when an event cannot occur i.e. there is no chance of the
event occurring it is said to be an impossible event. The probability of this event is
0.


#### Equally Likely Events

The events are said to be equally likely if all the possible outcomes occur with
same chance and there is no reason to expect any one of them to occur than one
other.

Example - during a throwing a dice, you are equally likely to get any number between 1-6


#### Mutually Exclusive Events

Two events are said to be mutually exclusive events when both cannot occur at
the same time. If the occurrence of one event rules out the possibility of occurrence of another event then the events are said to be Mutually Exclusive events. Mutually exclusive events always have a different outcome. Two
simple events are always mutually exclusive, whereas two compound events may
or may not be.


#### Independent and Dependent Events

Two or more events are said to be independent when the outcome of one does
not affect, and is not affected by the other.
Dependent events are those in which the occurrence or non-occurrence of one
event in any one trial affects the probability of other events in other trails.

In the below picture, suppose if a  person randomly chooses a black shoe , then he replaced it and choose a red shoe.  Since the first shoe was replaced, choosing a black shoe on the first try has no effect on the probability of choosing a red shoe on the second try. Therefore, these events are independent. But now if he has chose red shoe, he must chose the same shoe to make it a pair. Choosing a red shoe is affecting the on the probability that he will choose same shoe to make it pair. These events are dependent events.


#### Exhaustive Events


The events are said to be exhaustive if they include all the possible outcomes in
the sample space.
Ex  If we tossed two coins simultaneously, S= {HH, HT, TT, TH}. If the event is to
get a either a single head or tail, then the whole sample space is the outcome.


#### Favourable Events

The trial which corresponds to happening of an event is said to be favourable
events.
Ex  If we tossed two coins simultaneously, S= {HH, HT, TT, TH}. If the event is to
get a single head, then the outcomes favourable to it is HT or TH

The probability of a given event is an expression of likelihood or chance of
occurrence of an event. It is a concept to numerically measure the uncertainty
involved in the random experiment.

Given the sample space, the probability is a function which assigns a non negative
real number to every event A denoted by P(A) and is called as the probability of
event A provided it satisfies following three condition.The conditions are called as
the axioms
1. Axiom of Positivity - P(A) 0
2. Axiom of certainty  If A is a sure event then P(A)=1
3. Axiom of additivity  If A1,A2,...An are mutually exclusive events then
P(A1 U A2 U..U An) = P(A1) + P(A2)+... P(An)

 Where P(A) is the ratio of the number of favourable events (say m) to the total number of equally likely events (say n), hence P(A) = m/n.

### Bayes Theorem

The concept of conditional probability takes into account information about the occurence of one event to predict the probability of another event.  This concept can be extended to revise probabilities based on new information and to determine the probability that a particular effect was due to a specific cause. The procedure of revising these probabilities is known as Bayes theorem. 

Q While watching a game of IPL in a cafe, you observe someone who is clearly supporting Mumbai Indians in the game. What is the probability that they were actually born within 40kms of Mumbai ? Assume that: 
 the probability that a randomly selected person in a typical local cafe environment is born within 40 kms of Mumbai is 1/20 , and; 
 the chance that a person born within 40 kms of Mumbai actually supports Mumbai Indians is 7/10
  the probability that a person not born within  40 kms of Mumbai supports Mumbai Indians with probability 1/10 .

SOLUTION: 
Define 
B - event that the person is born within 40 kms of Mumbai
U - event that the person supports Mumbai Indians.

 We want P(B|U). By Bayes Theorem, 
P(B|U)  = P(U|B)P(B) /P(U)
 = P(U|B)P(B)/( P(U|B)P(B) + P(U|B )P(B ))
 = 7/10* 1/20/( 7/10 * 1/20 + 1/10 * 19/20) 
 = 7/26 
 0.269.


###  Bayes Theorem implementation from scratch ###

In layman's terms, the Bayes theorem calculates the probability of an event occurring, based on certain other probabilities that are related to the event in question. It is  composed of a  prior(the probabilities that we are aware of or that is given to us) and the posterior(the probabilities we are looking to compute using the priors). 

Let us implement the Bayes Theorem from scratch using a simple example. Let's say we are trying to find the odds of an individual having diabetes, given that he or she was tested for it and got a positive result. 
In the medical field, such probabilies play a very important role as it usually deals with life and death situations. 

We assume the following:

`P(D)` is the probability of a person having Diabetes. It's value is `0.01` or in other words, 1% of the general population has diabetes(Disclaimer: these values are assumptions and are not reflective of any medical study).

`P(Pos)` is the probability of getting a positive test result.

`P(Neg)` is the probability of getting a negative test result.

`P(Pos|D)` is the probability of getting a positive result on a test done for detecting diabetes, given that you have diabetes. This has a value `0.9`. In other words the test is correct 90% of the time. This is also called the Sensitivity or True Positive Rate.

`P(Neg|~D)` is the probability of getting a negative result on a test done for detecting diabetes, given that you do not have diabetes. This also has a value of `0.9` and is therefore correct, 90% of the time. This is also called the Specificity or True Negative Rate.

The Bayes formula is as follows:

<img src="bayes_formula.png" height="242" width="242">

* `P(A)` is the prior probability of A occurring independently. In our example this is `P(D)`. This value is given to us.

* `P(B)` is the prior probability of B occurring independently. In our example this is `P(Pos)`.

* `P(A|B)` is the posterior probability that A occurs given B. In our example this is `P(D|Pos)`. That is, **the probability of an individual having diabetes, given that, that individual got a positive test result. This is the value that we are looking to calculate.**

* `P(B|A)` is the likelihood probability of B occurring, given A. In our example this is `P(Pos|D)`. This value is given to us.

Putting our values into the formula for Bayes theorem we get:

`P(D|Pos) = P(D) * P(Pos|D) / P(Pos)`

The probability of getting a positive test result `P(Pos)` can be calculated using the Sensitivity and Specificity as follows:

`P(Pos) = [P(D) * Sensitivity] + [P(~D) * (1-Specificity))]`

```python
# P(D)
p_diabetes = 0.01

# P(~D)
p_no_diabetes = 0.99

# Sensitivity or P(Pos|D)
p_pos_diabetes = 0.9

# Specificity or P(Neg|~D)
p_neg_no_diabetes = 0.9

# P(Pos)
p_pos = (p_diabetes * p_pos_diabetes) + (p_no_diabetes * (1 - p_neg_no_diabetes))
print('The probability of getting a positive test result P(Pos) is: {}',format(p_pos))
```

**Using all of this information we can calculate our posteriors as follows:**
    
The probability of an individual having diabetes, given that, that individual got a positive test result:

`P(D|Pos) = (P(D) * Sensitivity)) / P(Pos)`

The probability of an individual not having diabetes, given that, that individual got a positive test result:

`P(~D|Pos) = (P(~D) * (1-Specificity)) / P(Pos)`

The sum of our posteriors will always equal `1`. 

```python

# P(D|Pos)
p_diabetes_pos = (p_diabetes * p_pos_diabetes) / p_pos
print('Probability of an individual having diabetes, given that that individual got a positive test result is:\
',format(p_diabetes_pos)) 
```

```python

# P(Pos|~D)
p_pos_no_diabetes = 0.1

# P(~D|Pos)
p_no_diabetes_pos = (p_no_diabetes * p_pos_no_diabetes) / p_pos
print('Probability of an individual not having diabetes, given that that individual got a positive test result is:'\
,p_no_diabetes_pos)
```
